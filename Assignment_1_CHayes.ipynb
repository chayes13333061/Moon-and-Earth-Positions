## Task 3: Data manipulation, Visualisation and Summarisation

Using dataframes created previously in task 2, we can manipulate the data accordingly. First define some useful methods.

```python
# Function that will take the dataframe as input and calculate the euclidean distance based on the x,y and z coordinates
# provided by the dictionary key-value pair. The dictionary key that contains the 3D coordinates is also passed as an
# argument. Returns an array of the euclidean distance of the 3D coordinates at each index of the dataframe.
def euclidean_distance_calc(dataframe, dict_key):
    out_array = []
    # Use pythagorus to calculate euclidean distance
    for i in range(len(dataframe)):
        out_array.append(math.sqrt(dataframe.iloc[i][str(dict_key)]["x"]**2 + dataframe.iloc[i][str(dict_key)]["y"]**2 + dataframe.iloc[i][str(dict_key)]["z"]**2))
    return out_array


# This function takes a dataframe as input and returns the x, y, and z coordinates of the celestial body based on the 
# dictionary key provided as input. This function returns 3 arrays of the x, y and z coordinates at each index in the
# dataframe.
def get_xyz_points(dataframe, dict_key):
    x = []
    y = []
    z = []
    # Retrieve x, y and z coordinates
    for i in range(len(dataframe)):
        x.append(dataframe.iloc[i][str(dict_key)]["x"])
        y.append(dataframe.iloc[i][str(dict_key)]["y"])
        z.append(dataframe.iloc[i][str(dict_key)]["z"]) 
    return x,y,z
```

Retrieve the date that measurements were taken. Determmine distance from earth using pythagorus. Retrieve x, y and z coordinates from dataframe.

```python
# Retrieve dates of lunar/earth distance measurements
lunar_dates = []
for i in range(len(total_lunar_dataframe)):
    lunar_dates.append(total_lunar_dataframe.iloc[i]["date"])

# Apply pythagorus to the x, y and z co-ordinates to give the euclidean distance to the moon.
# Store in lunar_distance array to be plotted later.
lunar_distances = euclidean_distance_calc(total_lunar_dataframe, "lunar_j2000_position")

# Store x, y and z distances from the moon to the earth in arrays
x_lunar_distances, y_lunar_distances, z_lunar_distances = get_xyz_points(total_lunar_dataframe, "lunar_j2000_position")
    
# Use pythagorus to determine the euclidean distance from the earth to the sun, and store in distance array
earth_distance_to_sun = euclidean_distance_calc(total_lunar_dataframe, "sun_j2000_position")

# Retrieve the distances in the x, y and z directions from the earth to the sun, store in distance arrays
x_earth_distances, y_earth_distances, z_earth_distances = get_xyz_points(total_lunar_dataframe, "sun_j2000_position")

# Check there are no missing data values
total_lunar_dataframe.isnull().sum()
```

### Summary of Earth and Moon Distances ###

```python
# Arrays to hold distance of moon from sun over year
x_distance_moon_to_sun = []
y_distance_moon_to_sun = []
z_distance_moon_to_sun = []

# Adding the earths distance to the moons distance we can get the total distance from moon to the sun.
# Append these values to the moon_to_sun distance array.
for i in range(len(x_lunar_distances)):
    x_distance_moon_to_sun.append(x_earth_distances[i] + x_lunar_distances[i])
    y_distance_moon_to_sun.append(y_earth_distances[i] + y_lunar_distances[i])
    z_distance_moon_to_sun.append(z_earth_distances[i] + z_lunar_distances[i]) 

# Calculate the euclidean distance from the moon to the sun over a year
moon_distance_to_sun = []
for i in range(len(x_distance_moon_to_sun)):
    moon_distance_to_sun.append(math.sqrt(x_distance_moon_to_sun[i]**2 + y_distance_moon_to_sun[i]**2 + z_distance_moon_to_sun[i]**2))
```

```python
# Plot both the earth and moon orbit around the sun over the year
plt.figure(9)
plt.plot(lunar_dates, earth_distance_to_sun, label="Earth")
plt.plot(lunar_dates, moon_distance_to_sun, label="Moon")
plt.title("Earth and Moon distance to Sun")
plt.xlabel("Month (yyyy-mm)")
plt.ylabel("Distance (km)")
plt.legend()
```

**Figure 9 (Earth and Moon Distance to Sun)**: In this plot, we can see that the earths distance to the sun varies over the year. The moon follows the orbit of the earth with oscillations around the earths orbit.
